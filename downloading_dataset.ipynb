```python
import os
import shutil

# Define source and destination directories
src_dir = "/content/data/new/nturgb+d_rgb"
dst_dir = "/content/data/filtered_videos"

# Create destination directory if it doesn't exist
os.makedirs(dst_dir, exist_ok=True)

# List of actions to keep
actions_to_keep = ["A001", "A003", "A006", "A011", "A012", "A022", "A027", "A028", "A032", "A038"]

# Iterate through files and move only the selected actions
for filename in os.listdir(src_dir):
    if any(action in filename for action in actions_to_keep):
        shutil.move(os.path.join(src_dir, filename), os.path.join(dst_dir, filename))

print("Selected videos moved successfully!")
```

```python
import os
import glob

# Define source directory
src_dir = "/content/data/filtered_videos"

# List of actions to check
actions =  ["A001", "A003", "A006", "A011", "A012", "A022", "A027", "A028", "A032", "A038"]

# Dictionary to store counts
action_counts = {action: 0 for action in actions}

# Get all .avi files in the folder
video_files = glob.glob(os.path.join(src_dir, "*.avi"))

# Iterate through files and count occurrences of each action
for file_path in video_files:
    filename = os.path.basename(file_path)  # Extract filename from full path
    for action in actions:
        if action in filename:
            action_counts[action] += 1

# Display counts
for action, count in action_counts.items():
    print(f"Total videos for {action}: {count}")
```

```python
import os
import shutil

# Define source directory
src_dir = "/content/data/filtered_videos"

# List of actions to organize
actions =  ["A001", "A003", "A006", "A011", "A012", "A022", "A027", "A028", "A032", "A038"]
# Create separate folders and move files
for action in actions:
    action_folder = os.path.join(src_dir, action)  # Folder for each action
    os.makedirs(action_folder, exist_ok=True)  # Create folder if it doesn't exist

    # Move matching videos into respective folders
    for filename in os.listdir(src_dir):
        if action in filename and filename.endswith(".avi"):  # Ensure only video files are moved
            shutil.move(os.path.join(src_dir, filename), os.path.join(action_folder, filename))

print("Videos have been organized into separate folders!")
```

```python
import os
import shutil

# Define the source and destination paths
source_dir = "/content/drive/MyDrive/NTU_Videos/A032/A032"
dest_dir = "/content/drive/MyDrive/NTU_Videos/A032"

# Move all content from source_dir to dest_dir
if os.path.exists(source_dir):
    for item in os.listdir(source_dir):
        source_item = os.path.join(source_dir, item)
        dest_item = os.path.join(dest_dir, item)

        # Move files and directories
        if os.path.isdir(source_item):
            shutil.move(source_item, dest_item)
        else:
            shutil.move(source_item, dest_item)
    print(f"All content from {source_dir} has been moved to {dest_dir}.")

    # Delete the empty source directory
    shutil.rmtree(source_dir)
    print(f"The directory {source_dir} has been deleted.")
else:
    print(f"The source directory {source_dir} does not exist.")
```
